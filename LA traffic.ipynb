So, the feature called "location" have latitude and longitude, we will need to work in that feature

```python
pd.option_context('display.max_rows', None, 'display.max_columns', None)
df2 = df['Location'].head()
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
#Let's check how the data is structured within each sample in the "Location" feature
df1['Location'][0]
```

```python
#extracting the data that we need

import re 
n=0
detail = '.'
detail2 = '-'

d = {'lat':[], 'long': []}
lat_long_df = pd.DataFrame(data=d)

while n<1000:
    data = re.findall(r'\d+',df['Location'][n])
    long = detail2+data[0]+detail+data[1]
    lat = data[2]+detail+data[3]
    
    finalData = {'lat':  lat,'long': long}
    lat_long_df = lat_long_df.append(finalData,ignore_index=True)
    n=n+1
    
lat_long_df
```

```python
lat_long_df.dtypes
```

```python
#Conveting data type
lat_long_df['lat'] = lat_long_df['lat'].astype(float)
lat_long_df['long'] = lat_long_df['long'].astype(float)
 
```

```python
lat_long_df.dtypes
```

```python
#Concatenating to our main df
df_new = pd.concat([df1, lat_long_df], axis=1)
```

```python
import folium  #geospatial tools
from IPython.display import HTML
from folium.plugins import HeatMap
```

```python
#Get LA coordinates from google and define our map
map_la = folium.Map(location=(34.052235,-118.243683))
```

```python
map_la
```

```python
#Loop to fill the map with ours coordinates from "df_new"
#Using heat map

data = []
for i in range(len(df_new)):
    data.append((df_new['lat'][i],df_new['long'][i]))
HeatMap(data,radius=9).add_to(map_la)
map_la.save('la_accidents.html')
```

```python
#Defining how HTML it's going to be saved
HTML(r'<iframe width="800" height="500" frameborder="0" allowfullscreen src="./la_accidents.html"></iframe>')
```

```python
#Using circle marks

map_la2 = folium.Map(location=(34.052235,-118.243683))

# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, long, in zip(df_new['lat'], df_new['long']):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat, long],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
map_la2.add_child(incidents)
```

```python
from folium import plugins

#Using plugins

# let's start again with a clean copy of the map of LA
map_la3 = folium.Map(location=(34.052235,-118.243683))

# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(map_la3)

# loop through the dataframe and add each data point to the mark cluster
for lat, long, label, in zip(df_new['lat'], df_new['long'], df_new['Crime Code Description']):
    folium.Marker(
        location=[lat, long],
        icon=None,
        popup=label,
    ).add_to(incidents)

# display map
map_la3
```
